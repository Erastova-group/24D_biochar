```python
import MDAnalysis as mda
from MDAnalysis.lib import distances
from MDAnalysis.lib import mdamath
from MDAnalysis.lib.log import ProgressBar

import numpy as np
import pandas as pd
import os

import matplotlib.pyplot as plt
import matplotlib

import pickle
import scipy
```

## list systems to analyze

```python
systems=[]

for BC in ['BC400/', 'BC600/','BC800/']:
    for c in ['c1/','c2/','c3/','c4/', 'c5/','c6/','c7/','c8/']:
        systems.append(BC+c)
        
print(systems)
```

## load universes, calculate angles and save to pickle

**note: skip to next section if you've already pickled your angles**

```python
# check whether you've already pickled your angles
if os.path.exists('total_angles.pickle'):
    print('Pickle exists already!')
```

```python
def load_universe():
    
    # load universe coords and trajectory (u)
    u=mda.Universe('system.top', 'eq_sys_pbc.xtc', topology_format='ITP')
    # add extra info into universe (u)
    u.add_TopologyAttr('elements', [re.split(r'(\d+)', a)[0] for a in u.atoms.names])
    print('loaded universe',flush=True)
    
    return u
```

```python
def find_angles(res):
    
    # identify atoms to form surface plane
    res_atoms=u.select_atoms(f"resnum {str(res)} and type opls_800 opls_801 opls_802 opls_803 opls_804 opls_805")
    com_x, com_y, com_z=res_atoms.atoms.center_of_mass()
    
    # indentify close interfacial atoms (ias) - within 4.2 Å to com of aromatic ring
    cut_off = 4.2
    close_ias=biochar.atoms.select_atoms(f"point {str(com_x)} {str(com_y)} {str(com_z)} {cut_off} and name C*", updating=True)
    close_ias_ids=close_ias.ids
    
    # check if there are at least 3 surface atoms within 4.2 Å of res
    if len(close_ias)>=3: 
        
        # calculate distances between res and close_ias, sort and identify closest three surf atoms
        dists=distances.distance_array(close_ias.positions, res_atoms.atoms.center_of_mass())
        sorted_dists,sorted_close_ias_ids = zip(*sorted(zip(dists,close_ias_ids)))
        s0, s1, s2 = close_ias.select_atoms(f"id {sorted_close_ias_ids[0]} {sorted_close_ias_ids[1]} {sorted_close_ias_ids[2]}")
        
        # check surface atoms are on same residue
        if s0.resid == s1.resid or s0.resid == s2.resid: 
            
            # identify res plane + normal
            a0, a1, a2 = res_atoms.atoms.select_atoms('type opls_800 opls_802 opls_804')
            positions_a0 = a0.position
            positions_a1 = a1.position
            positions_a2 = a2.position
            
            vector_a1=positions_a1 - positions_a0
            vector_a2=positions_a2 - positions_a0
            normal_res=mdamath.normal(vector_a1,vector_a2)
            
            # identify surface plane + normal
            positions_s0 = s0.position
            positions_s1 = s1.position
            positions_s2 = s2.position
            
            vector_s1=positions_s1 - positions_s0
            vector_s2=positions_s2 - positions_s0
            normal_surf=mdamath.normal(vector_s1,vector_s2)
            
            # calculate angle between two normal vectors
            angle_res_surf=mdamath.angle(normal_res,normal_surf)*180/np.pi
            
            if angle_res_surf>90:
                angle_res_surf=abs(180-angle_res_surf)
        
            return angle_res_surf
    
    return np.nan
```

```python
total_angles={}

for system in ProgressBar(systems):
    
    u = load_universe(system)
    
    # identify biochar atoms
    biochar = u.select_atoms('resname ?G*')

    # identify 2,4-D (DCPAA) atoms and resids
    DCPAA_total=u.select_atoms('resname DCPAA')
    DCPAA_residues=u.select_atoms('resname DCPAA').residues.resids
    
    angles=[]
    
    # check angles every 10 steps in trajectory
    for ts in u.trajectory[-2000::10]:
        for res in DCPAA_residues:
            angle_res_surf = find_angles(res)
            angles.append(angle_res_surf)
            
    # save all angles (all residues, all timesteps) for each system and print mean + std
    total_angles[system]=angles
    print(system,'\nmean:',np.nanmean(total_angles[system]),'std:',np.nanstd(total_angles[system]))
```

```python
# save angles_total (dictionary) as pickle to save time later
with open('total_angles.pickle', 'wb') as f:
    pickle.dump(total_angles, f, protocol=pickle.HIGHEST_PROTOCOL)
```

## load from pickle and plot

**note: skip to here if you've already pickled your angles**

```python
#load angles_total (dictionary) from pickle if already calculated

with open('total_angles.pickle', 'rb') as f:
    total_angles=pickle.load(f)
```

```python
matplotlib.rcParams.update({'font.size':14, 'figure.autolayout':True})

hist={}
bin_edges={}

# plot as histogram and save
for system in mda.lib.log.ProgressBar(systems):
    hist[system], bin_edges[system] = np.histogram(total_angles[system], bins=90, range=(0,90), density=True)
    
    plt.hist(total_angles[system], bins=90, density=True, color='g', edgecolor='k')
    plt.ylim(0,0.1)
    plt.ylabel('Probability density')
    plt.xlabel('Angle between aromatic ring and biochar surface /º')
    plt.savefig('figures/fig_'+system[:5]+system[6:8]+'_angles.png')
    plt.show()
    plt.clf()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=[8,3*6])

#comparison between different systems
for i, system in enumerate(systems):
    if system.startswith('BC400'):
        axs[0].bar(bin_edges[system][:-1]+0.5,hist[system],align='center',label=system)
        
    if system.startswith('BC600'):
        axs[1].bar(bin_edges[system][:-1]+0.5,hist[system],align='center',label=system)

    if system.startswith('BC800'):
        axs[2].bar(bin_edges[system][:-1]+0.5,hist[system],align='center',label=system)

axs[0].legend()
axs[1].legend()
axs[2].legend()
```
