## get fnames

```python
systems=[]

for BC in ['BC400/', 'BC600/','BC800/']:
    for c in ['c1/','c2/','c3/','c4/', 'c5/','c6/','c7/']:
        systems.append(BC+c)
```

```python
fnames_ar=[]
fnames_acidOs=[]
fnames_cl6=[]
fnames_cl7=[]
fnames_etherOs=[]
fnames_NA=[]

for system in systems:
    fnames_ar.append(system+'closest_atoms_aromatics_df.csv')
    fnames_acidOs.append(system+'closest_atoms_acidOs_df.csv')
    fnames_cl6.append(system+'closest_atoms_chlorine6_df.csv')
    fnames_cl7.append(system+'closest_atoms_chlorine7_df.csv')
    fnames_etherOs.append(system+'closest_atoms_etherOs_df.csv')
    fnames_NA.append(system+'closest_atoms_NA_df.csv')
```

```python
fnames_tot=fnames_ar+fnames_cl6+fnames_cl7+fnames_etherOs+fnames_acidOs+fnames_NA
```

## load data

```python
import pandas as pd
import numpy as np
```

```python
df={}
data={}

for i, f in enumerate(fnames_tot):
    #read in data
    df[f]=pd.read_csv(f, index_col=0)
    data[f]=df[f].values.ravel().reshape(-1,1)
```

## set up universes and tally up elements

```python
import MDAnalysis as mda
import re
import matplotlib.pyplot as plt
```

```python
def load_universe(system):
    
    # load universe coords and trajectory (u)
    u=mda.Universe(system+'system.top', topology_format='ITP')
    # add extra info into universe (u)
    u.add_TopologyAttr('elements', [re.split(r'(\d+)', a)[0] for a in u.atoms.names])
    print('loaded universe',flush=True)
    
    return u
```

```python
for system in systems:
    
    u=load_universe(system)
    
    for f in fnames_tot:
        if f.startswith(system)==False:
            pass
        
        else:
            C, H, O = 0, 0, 0
            
            for i in data[f]:
                if u.atoms[i].elements == 'C':
                    C+=1
                elif u.atoms[i].elements == 'H':
                    H+=1
                elif u.atoms[i].elements == 'O':
                    O+=1
            
            closest_elements_dic[system][f]=[C,H,O]

        
        pass
```
