## get fnames

```python
systems=[]

for BC in ['BC400/', 'BC600/','BC800/']:
    for c in ['c1/','c2/','c3/','c4/', 'c5/','c6/','c7/']:
        systems.append(BC+c)
```

```python
fnames_ar=[]
fnames_acidOs=[]
fnames_cl6=[]
fnames_cl7=[]
fnames_etherOs=[]
fnames_NA=[]

for system in systems:
    fnames_ar.append(system+'dists_aromatics_df.csv')
    fnames_acidOs.append(system+'dists_acidOs_df.csv')
    fnames_cl6.append(system+'dists_chlorine6_df.csv')
    fnames_cl7.append(system+'dists_chlorine7_df.csv')
    fnames_etherOs.append(system+'dists_etherOs_df.csv')
    fnames_NA.append(system+'dists_NA_df.csv')
```

```python
fnames_tot=fnames_ar+fnames_cl6+fnames_cl7+fnames_etherOs+fnames_acidOs+fnames_NA
```

## load data

```python
import pandas as pd
import numpy as np
```

```python
df={}
data={}
hist={}
bin_edges={}

for i, f in enumerate(fnames_tot):
    #read in data
    df[f]=pd.read_csv(f, index_col=0)
    data[f]=df[f].values.ravel().reshape(-1,1)
    hist[f], bin_edges[f]=np.histogram(data[f],bins=100,range=(2,15),density=True)
    hist[f]=hist[f]*len(df[f].columns)
```

## get no. of adsorbed residues
use aromatic part to guage this

```python
avg_no_adsorbed_residues={}
avg_no_free_residues={}

percent_adsorbed_residues={}
percent_free_residues={}

for f in fnames_ar:
    threshold=4.2
    
    no_adsorbed_residues=dict((df[f]<=threshold).sum(axis=1))
    no_free_residues=dict((df[f]>threshold).sum(axis=1))
    avg_no_adsorbed_residues[f]=np.mean(list(no_adsorbed_residues.values()))
    avg_no_free_residues[f]=np.mean(list(no_free_residues.values()))
    percent_adsorbed_residues[f]=avg_no_adsorbed_residues[f]/len(df[f].columns)*100
    percent_free_residues[f]=avg_no_free_residues[f]/len(df[f].columns)*100
```

## generate adsorption isotherms

use mda to get no. water residues in system, calc eq_conc and adsorbed_amount 

```python
import MDAnalysis as mda
from scipy.constants import N_A, R
```

```python
#constants
dens_water=3.34563E+28 #molecules m-3 of water
SASA={'BC400':1.62374E-16, 'BC600':2.33353E-16, 'BC800':7.78946E-16} # m2 surface area 
SSA={'BC400':100, 'BC600':150, 'BC800':200} # m2g-1 surface area 

DCPAA_mw=220.03 # g mol-1
```

```python
L_sol={}
eq_conc={} # mol L-1
eq_conc_g={} # g L-1
adsorbed_amount={} # mol m-2
adsorbed_amount_pg={} # mg g-1

BC400_data=[]
BC600_data=[]
BC800_data=[]

for system in mda.lib.log.ProgressBar(systems):
    u=mda.Universe(system+'eq_sys.gro')
    sol=u.select_atoms('resname SOL')
    n_sol=sol.residues.n_residues
    L_sol[system]=n_sol/dens_water*1000
    eq_conc[system]=avg_no_free_residues[system+'dists_aromatics_df.csv']/N_A/L_sol[system]
    eq_conc_g[system]=eq_conc[system]*DCPAA_mw
    adsorbed_amount[system]=avg_no_adsorbed_residues[system+'dists_aromatics_df.csv']*1E6/(N_A*SASA[system[:5]])
    adsorbed_amount_pg[system]=(adsorbed_amount[system]*DCPAA_mw)*SSA[system[:5]]/1E3
    
    if system.startswith('BC400'):
        BC400_data.append([eq_conc[system],eq_conc_g[system],adsorbed_amount[system],adsorbed_amount_pg[system]])
    if system.startswith('BC600'):
        BC600_data.append([eq_conc[system],eq_conc_g[system],adsorbed_amount[system],adsorbed_amount_pg[system]])
    if system.startswith('BC800'):
        BC800_data.append([eq_conc[system],eq_conc_g[system],adsorbed_amount[system],adsorbed_amount_pg[system]])
```

```python
BC400_df=pd.DataFrame(BC400_data, columns=['eq conc','eq conc g','adsorbed amount', 'adsorbed amount pg'])
BC600_df=pd.DataFrame(BC600_data, columns=['eq conc','eq conc g','adsorbed amount', 'adsorbed amount pg'])
BC800_df=pd.DataFrame(BC800_data, columns=['eq conc','eq conc g','adsorbed amount', 'adsorbed amount pg'])
```

## fit langmuir curves

```python
from scipy.optimize import curve_fit
from scipy.constants import R
```

```python
def dilangmuir(ce, qm1=20, k1=1, qm2=10, k2=1):
    q=(qm1*k1*ce)/(1+k1*ce) + (qm2*k2*ce)/(1+k2*ce)
    return q

langmuir_bounds=((0,0,0,0),(np.inf,np.inf,np.inf,np.inf))
```

### per gram

```python
def plot_dilangmuir_pg(x, y):
    
    popt, pcov = curve_fit(dilangmuir, x, y, bounds=langmuir_bounds)
    print(popt)
    
    fig = plt.figure()
    plt.scatter(x,y, marker='s',color='k')
    plt.plot(np.linspace(0,400),dilangmuir(np.linspace(0,400),*popt),'b--')
    plt.xlabel('Equilibrium 2,4-D concentration (g L$^{-1}$)')
    plt.ylabel('2,4-D adsorbed amount (mg g$^{-1}$)')
    
    return popt, fig
```

```python
BC400_popt, fig = plot_dilangmuir_pg(BC400_df['eq conc g'], BC400_df['adsorbed amount pg'])
fig.savefig('figures/BC400_isotherm_pg.png')
```

```python
BC600_popt, fig = plot_dilangmuir_pg(BC600_df['eq conc g'],BC600_df['adsorbed amount pg'])
fig.savefig('figures/BC600_isotherm_pg.png')
```

```python
BC800_popt, fig = plot_dilangmuir_pg(BC800_df['eq conc g'], BC800_df['adsorbed amount pg'])
fig.savefig('figures/BC800_isotherm_pg.png')
```

```python
x_linspace=np.linspace(0,400)

plt.scatter(BC400_df['eq conc g'],BC400_df['adsorbed amount pg'],marker='s',color='b', label='BC400')
plt.plot(x_linspace,dilangmuir(x_linspace,*BC400_popt),'b--')
plt.scatter(BC600_df['eq conc g'],BC600_df['adsorbed amount pg'],marker='s',color='g', label='BC600')
plt.plot(x_linspace,dilangmuir(x_linspace,*BC600_popt),'g--')
plt.scatter(BC800_df['eq conc g'],BC800_df['adsorbed amount pg'],marker='s',color='r', label='BC800')
plt.plot(x_linspace,dilangmuir(x_linspace,*BC800_popt),'r--')
plt.xlabel('Equilibrium 2,4-D concentration (g L$^{-1}$)')
plt.ylabel('2,4-D adsorbed amount (mg g$^{-1}$)')
plt.legend()
plt.savefig('figures/BC_isotherm_pg.png')
plt.show()
```

### calc properties

```python
def calc_props_pm(x, y):
    
    popt, pcov = curve_fit(dilangmuir, x, y, bounds=langmuir_bounds)
    qm1, k1, qm2, k2 = popt
    
    #A = avg. area per molecule
    Am1=1E20/(N_A*(qm1)*1E-6)
    Am2=1E20/(N_A*(qm2)*1E-6)
    
    #dG = free energy to adsorb 1 mol of DCPAA
    c_water=(1E3/18)
    dG1=-R*300*np.log((k1)*c_water)
    dG2=-R*300*np.log((k2)*c_water)

    return Am1, Am2, dG1, dG2
```

```python
#for Am and dG calc use molar units
Am1, Am2, dG1, dG2 = calc_props_pm(BC400_df['eq conc'], BC400_df['adsorbed amount'])

print('BC400 qm =',BC400_popt[0]+BC400_popt[2],'mg g-1 \nBC400 A =',Am1,Am2,'2 molecule-1 \nBC400 dG =',dG1/1000,dG2/1000,'kJ mol-1')
```

```python
#for Am and dG calc use molar units
Am1, Am2, dG1, dG2 = calc_props_pm(BC600_df['eq conc'], BC600_df['adsorbed amount'])

print('BC600 qm =',BC600_popt[0]+BC600_popt[2],'mg g-1 \nBC600 A =',Am1,Am2,'2 molecule-1 \nBC600 dG =',dG1/1000,dG2/1000,'kJ mol-1')
```

```python
#for Am and dG calc use molar units
Am1, Am2, dG1, dG2 = calc_props_pm(BC800_df['eq conc'], BC800_df['adsorbed amount'])

print('BC800 qm =',BC800_popt[0]+BC800_popt[2],'mg g-1 \nBC800 A =',Am1,Am2,'2 molecule-1 \nBC800 dG =',dG1/1000,dG2/1000,'kJ mol-1')
```

## interpolate data

```python
from scipy.interpolate import splev, splrep
```

```python
spl_x={}
spl_y={}
dy={}
d2y={}

crossings={f:[] for f in fnames_cl}
crossings2={f:[] for f in fnames_cl}


fig, axs = plt.subplots(nrows=len(fnames_cl), ncols=1, figsize=[8,len(fnames_cl)*6])

for i, f in enumerate(fnames_cl):
    #define x and y from hist data
    x=bin_edges[f][:-1]+bin_widths
    y=hist[f]
    
    #find the spline representation
    spl=splrep(x, y, s=0.01, per=False)
    
    #define new x and y data using larger no. of points
    spl_x[f] = np.linspace(x[0], 6, 200)
    spl_y[f] = splev(spl_x[f], spl)
    
    dy[f]=splev(spl_x[f], spl, der=1)
    d2y[f]=splev(spl_x[f], spl, der=2)

    if dy[f][0] > 0 :
        pos=True

    for j, k in enumerate(dy[f]):
        if pos==True and  k < 0:
            crossings[f].append(j)
            pos=False
        if pos==False and k > 0:
            crossings[f].append(j)
            pos=True
    
    if d2y[f][0] > 0 :
        pos=True

    for j, k in enumerate(d2y[f]):
        if pos==True and  k < 0:
            crossings2[f].append(j)
            pos=False
        if pos==False and k > 0:
            crossings2[f].append(j)
            pos=True
    
    axs[i].plot(spl_x[f],spl_y[f])
    
    for crossing in crossings[f]:
        axs[i].plot(spl_x[f][crossing], spl_y[f][crossing], label=spl_x[f][crossing], color='b', marker='x')
        
    for crossing in crossings2[f]:
        axs[i].plot(spl_x[f][crossing], spl_y[f][crossing], label=spl_x[f][crossing], color='r', marker='x')
    
    axs[i].legend()
```
